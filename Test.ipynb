```python
player_list = players.get_active_players()
player_id_list = []
for pl in player_list:
    player_id_list.append(pl['id'])
```

```python
p = playercareerstats.PlayerCareerStats(player_list[0]['id']).season_totals_regular_season
p.get_dict()
```

```python
t = commonplayerinfo.CommonPlayerInfo(player_list[0]['id']).player_headline_stats
t.get_dict()['data'][0]
```

```python
player_data = {}
for i, pid in enumerate(player_id_list):
    p = commonplayerinfo.CommonPlayerInfo(pid).player_headline_stats
    try:
        s = p.get_dict()['data'][0]
        player_data[s[1]] = s
    except:
        pass
    print(i + 1, 'of', len(player_id_list))
```

```python
for key in player_data.keys():
    player_data[key] = player_data[key][1:-1]
```
